# HIGHEST GROSSING INDIAN MOVIES BY YEAR

```python
#IMPORTING PANDAS, REQUESTS AND BEATIFULSOUP
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#PASSING SITE URL IN A VARIABLE
url='https://juksun.com/list-of-highest-grossing-indian-movies-by-year/'
```

```python
#USING REQUESTS TO GET URL RESPONSE 
page=requests.get(url)
page
```

```python
#USING BEAUTIFULSOUP AND PRETTIFY TO GET HTML CODE
soup=BeautifulSoup(page.text)
soup.prettify
```

```python
#TAKING TABLE ID FROM HTML IN A VARIABLE
table=soup.find('article',id='post-23028')
```

```python
table
```

```python
#TAKING TABLE BODY FROM HTML IN A VARIABLE
tbody=table.find('tbody')
```

```python
#TAKING TABLE ROWS FROM HTML IN A VARIABLE
rows=tbody.find_all('tr',style='')
rows
```

```python
#TAKING TABLE DATA FROM ROWS USING LOOPS 
column_info=[]
for i in rows:
    cols=i.find_all('td')
    book_info=[c.text for c in cols]
    column_info.append(book_info)
print(column_info)
```

```python
#FOR TAKING TABLE DATA UNDER IT'S RESPECTIVE COULMNS
Year=[]
Title=[]
Director=[]
Production_Company=[]
Worldwide_gross_revenue=[]
Inflation_adjusted_gross=[]
Footfalls=[]
Original_language=[]
for i in column_info:
    Year.append(i[0])
    Title.append(i[1])
    Director.append(i[2])
    Production_Company.append(i[3])
    Worldwide_gross_revenue.append(i[4])
    Inflation_adjusted_gross.append(i[5])
    Footfalls.append(i[6])
    Original_language.append(i[7])
    
```

```python
Data={'Year':Year,'Title':Title,'Director':Director,'Production_Company':Production_Company,'Worldwide gross revenue':Worldwide_gross_revenue,'Inflation adjusted gross':Inflation_adjusted_gross,'Footfalls':Footfalls,'Original language':Original_language}
```

```python
#CREATING PANDAS DATAFRAME 
movie_info=pd.DataFrame(Data)
movie_info
```

```python
#CREATING .CSV FILE FROM DATAFRAME
movie_info.to_csv('movie.csv')
```
